```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python

sns.violinplot(x='species', y='petal_length', data=iris)
plt.title('Violin Plot')
plt.xticks(color='grey')
plt.show()


```

```python
sns.stripplot(x='species', y='petal_length', data=iris)
plt.title('Strip Plot')
plt.xticks(color='red')
plt.show()
```

```python
sns.swarmplot(x='species', y='petal_length', data=iris, size=3)
plt.xticks(color='red')
plt.title('Swarm Plot')
plt.show()
```

```python
sns.boxplot(x='species', y='petal_length', data=iris,color='maroon')
plt.title('Box Plot')
plt.xticks(color="grey")
plt.show()
```

```python
sns.barplot(x='species', y='petal_length', data=iris,)
plt.title('Bar Plot')
plt.xticks(color='grey')
plt.show()
```

```python
sns.countplot(x='species', data=iris,color='lightgreen')
plt.title('Count Plot')
plt.xticks(color='grey')
plt.show()
```

```python
numerical_iris = iris.select_dtypes(include=['float', 'int'])
corr_matrix = numerical_iris.corr()
sns.heatmap(numerical_iris.corr(), annot=True)
plt.title('Heatmap')
plt.show()
```

```python
sns.scatterplot(x='petal_length', y='petal_width', data=iris,color='red')
plt.title('Scatter Plot')
plt.show()
```

```python
sns.lineplot(iris)
plt.title('Line Plot')
plt.show()
```
