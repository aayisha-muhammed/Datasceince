```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# Load the dataset
house_data = pd.read_csv('House_Pricing.csv')

# Display the first few rows to understand the dataset
house_data.head()
```

```python
# Select numeric columns
numeric_columns = house_data.select_dtypes(include=[np.number]).columns

# Fill missing numerical values with the median
house_data[numeric_columns] = house_data[numeric_columns].apply(lambda x: x.fillna(x.median()))

# Check for remaining missing values
print(house_data.isnull().sum())
```

```python
# Convert the date column to datetime
house_data['Date House was Sold'] = pd.to_datetime(house_data['Date House was Sold'], errors='coerce')

# Check if conversion was successful
print(house_data['Date House was Sold'].head())
```

```python
# Convert the date column to datetime
house_data['Date House was Sold'] = pd.to_datetime(house_data['Date House was Sold'], errors='coerce')

# Check if conversion was successful
print(house_data['Date House was Sold'].head())
```

```python
# Calculate House Age from "Date House was Sold" and "Age of House (in Years)"
house_data['House Age'] = house_data['Date House was Sold'].dt.year - (2024 - house_data['Age of House (in Years)'])

# Display the newly created column
print(house_data['House Age'].head())
```

```python
# Drop unnecessary columns like 'ID' and the original date column
house_data.drop(['ID', 'Date House was Sold'], axis=1, inplace=True)

# Display the updated DataFrame
print(house_data.head())
```

```python
# Remove 'ID' from numeric_columns
numeric_columns = numeric_columns.drop('ID')

# Initialize the scaler
scaler = StandardScaler()

# Apply scaling to numeric columns
house_data[numeric_columns] = scaler.fit_transform(house_data[numeric_columns])

# Display the scaled DataFrame
print(house_data.head())
```

```python
# Save the processed data to a new CSV file
house_data.to_csv('Processed_House_Pricing.csv', index=False)
```
